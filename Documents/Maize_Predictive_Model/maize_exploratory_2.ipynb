```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfs = pd.read_csv('Desktop/maize_data.csv')
```

```python
dfs.columns
```

```python
dfs.describe()
```

```python
dfs.info()
```

```python
dfs['Year'] = pd.to_datetime(dfs['Year'])
```

## Begin by analyzing various sources of Supply
   <ul>
   <li> Area Harvested </li>
   <li> Production </li>
   <li> Imports </li>
   <li> Yield/ Hectare </li>
   </ul>

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# Use ax1 to plot World
dfs[dfs['Country'] == 'World'].plot(x='Year', y='Area_Harvested', legend=False, ax=ax1)
ax1.set_title("Area Harvested - World")

# Use ax2 to plot Nigeria
dfs[dfs['Country'] == 'Nigeria'].plot(x='Year', y='Area_Harvested', legend=False, ax=ax2)
ax2.set_title("Area Harvested - Nigeria")

# If you don't do tight_layout() you'll have weird overlaps
plt.tight_layout()
plt.style.use('classic')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# Use ax1 to plot World
dfs[dfs['Country'] == 'World'].plot(x='Year', y='Production', legend=False, ax=ax1, style='c-')
ax1.set_title("Production - World")

# Use ax2 to plot Nigeria
dfs[dfs['Country'] == 'Nigeria'].plot(x='Year', y='Production', legend=False, ax=ax2, style='c-')
ax2.set_title("Production - Nigeria")

# If you don't do tight_layout() you'll have weird overlaps
plt.tight_layout()
plt.style.use('seaborn')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# Use ax1 to plot World
dfs[dfs['Country'] == 'World'].plot(x='Year', y='Imports', legend=False, ax=ax1, style='k--')
ax1.set_title("Imports - World")

# Use ax2 to plot Nigeria
dfs[dfs['Country'] == 'Nigeria'].plot(x='Year', y='Imports', legend=False, ax=ax2, style='k--')
ax2.set_title("Imports - Nigeria")

# If you don't do tight_layout() you'll have weird overlaps
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5), sharex=True, sharey=True)

# Use ax1 to plot World
dfs[dfs['Country'] == 'World'].plot(x='Year', y='Yield', legend=False, ax=ax1, style='g--')
ax1.set_title("Yield/Hectare - World")

# Use ax2 to plot Nigeria
dfs[dfs['Country'] == 'Nigeria'].plot(x='Year', y='Yield', legend=False, ax=ax2, style='g--')
ax2.set_title("Yield/Hectare - Nigeria")

# If you don't do tight_layout() you'll have weird overlaps
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# Use ax1 to plot World
dfs[dfs['Country'] == 'World'].plot(x='Year', y='Total_Supply', legend=False, ax=ax1)
ax1.set_title("Total Supply - World")

# Use ax2 to plot Nigeria
dfs[dfs['Country'] == 'Nigeria'].plot(x='Year', y='Total_Supply', legend=False, ax=ax2)
ax2.set_title("Total Supply - Nigeria")

# If you don't do tight_layout() you'll have weird overlaps
plt.tight_layout()
plt.style.use('classic')
plt.show()
```

### From the above charts we can draw the following conclusions about Nigerian Maize Supply

<ol> 
<li> Nigeria's Area Harvested has declined since 2014, so has overall Production.However, worldwide Area under harvest and Production has increased worldwide </li>
<li> Yield/Hectare has stayed low for Nigeria, while increasing worldwide </li>
<li> Nigerian Maize imports have doubled since 2013, presumably to keep up with demand </li>
</ol>

```python
dfn = dfs[dfs.Country == 'Nigeria']
```

```python
dfn = dfn.loc[:,['Year','Domestic_Consumption','Total_Supply']]
```

```python
dfn ['Demand_yoy'] =dfn.Domestic_Consumption.pct_change()
dfn ['Supply_yoy'] =dfn.Total_Supply.pct_change()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# Use ax1 to plot World
dfn.plot(x='Year', y='Demand_yoy', legend=False, ax=ax1)
ax1.set_title("Demand Change YoY - Nigeria")

# Use ax2 to plot Nigeria
dfn.plot(x='Year', y='Supply_yoy', legend=False, ax=ax2)
ax2.set_title("Supply Change YoY - Nigeria")

# If you don't do tight_layout() you'll have weird overlaps
plt.tight_layout()
plt.show()
```

### From the above charts we can draw the following conclusions about Nigerian Maize Demand

<ol> 
<li> Nigeria's Food demand for maize peaked in 2012 but has since been declining. World maize demand has seen a steady increase. </li>
<li> The Poultry sector accounts for 10-15% of all Maize demand in Nigeria, that demand has stayed strong </li>
<li> As a result, Overall Maize demand has experienced a steep decline in recent years. </li>
</ol>
